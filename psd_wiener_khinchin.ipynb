# Demo using np.fft.fft to calculate power spectrum

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from disp import set_font_size
```

Frequency array same length as time-series.

```python
dt = .001  # timestep size
fs = 1/dt  # sampling freq
t = np.arange(1, 3, dt)  # time vector
# noisy signal
x = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*50*t) + .5*np.sin(2*np.pi*300*t) + np.random.normal(0, .1, len(t))

n = len(x)
fft = np.fft.fft(x, n)
psd = np.abs(fft)**2
frq = np.fft.fftfreq(n, dt)

print('len(frq) = ', len(frq))

fig, axs = plt.subplots(1, 2, figsize=(10, 3), tight_layout=True)
axs[0].plot(t, x)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Signal')
axs[0].set_title('Signal')

axs[1].plot(frq, psd)
axs[1].set_xlim(0, fs/2)
axs[1].set_xlabel('Freq (Hz)')
axs[1].set_ylabel('Power')

for ax in axs:
    set_font_size(ax, 16)
```

Frequency array longer than time-series

```python
# n > len(t)
dt = .001
fs = 1/dt
t = np.arange(1, 3, dt)
x = np.sin(2*np.pi*5*t) + np.sin(2*np.pi*50*t) + .5*np.sin(2*np.pi*300*t) + np.random.normal(0, .1, len(t))

n = int(np.round(1.2*len(x)))

fft = np.fft.fft(x, n)
psd = np.abs(fft)**2
frq = np.fft.fftfreq(n, dt)

print('len(frq) = ', len(frq))

fig, axs = plt.subplots(1, 2, figsize=(10, 3), tight_layout=True)
axs[0].plot(t, x)
axs[0].set_xlabel('Time (s)')
axs[0].set_ylabel('Signal')
axs[0].set_title('Signal')

axs[1].plot(frq, psd)
axs[1].set_xlim(0, fs/2)
axs[1].set_xlabel('Freq (Hz)')
axs[1].set_ylabel('Power')

for ax in axs:
    set_font_size(ax, 16)
```

# Power spectral density and autocorrelation of random process

```python
%reset -f
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from disp import set_font_size


def cor_psd(xs, ys, dt, scale):
    '''
    Return (cross/auto) correlation and (cross/auto) power spectral density estimated
    from a set of signals.
    Each row of xs or ys is a 1-D trial. ys = xs for autocorr.
    
    scale = True divides by product of xs and ys std and gives 1 for max autocorrelation of signal
    
    To do: optional mean-subtraction to get proper autocovariance.
    '''
    n_t = xs.shape[1]
    frq = np.fft.fftfreq(n_t, dt) % (1/dt)
    
    ffts_x = np.fft.fft(xs, axis=1) / np.sqrt(n_t)
    ffts_y = np.fft.fft(ys, axis=1) / np.sqrt(n_t)
    psd = np.mean(ffts_x * np.conj(ffts_y), axis=0)
    
    cor = np.real(np.fft.ifft(psd))
    
    if scale:
        cor /= (xs.std() * ys.std())
    
    cut = int(np.floor(n_t/2))
    cor = np.concatenate([cor[cut:], cor[:cut]])
    tc = np.concatenate([np.arange(cut-n_t, 0.)*dt, np.arange(0., cut)*dt])
    
    return cor, tc, psd, frq
```

```python
# test on 1-D driven exponential filter
FPS = 30
DT = 1/FPS
np.random.seed(0)
t = np.arange(0, 300, DT)
xs = np.nan * np.zeros((5, len(t)))
tau = .5

for ctr in range(len(xs)):
    xs[ctr, 0] = 0
    for ct, _t in enumerate(t[1:], 1):
        dx = (DT/tau) * (-xs[ctr, ct-1] + .1 * np.random.randn())
        xs[ctr, ct] = xs[ctr, ct-1] + dx
    
acor, tac, psd, frq = cor_psd(xs, xs, DT, scale=True)

fig, axs = plt.subplots(1, 3, figsize=(14, 3.5), tight_layout=True)
axs[0].plot(t, xs[0, :])
axs[0].plot(t, xs[1, :])
axs[0].set_xlim(100, 110)
axs[0].set_xlabel('t (s)')
axs[0].set_ylabel('x')
axs[0].set_title('Signal')
axs[0].legend(['Sample 1', 'Sample 2'])

axs[1].plot(frq, psd, c='k')
axs[1].set_xlabel('Freq (Hz)')
axs[1].set_ylabel('Power')
axs[1].set_title('Power spectral density')

axs[2].plot(tac, np.exp(-np.abs(tac)/tau), lw=2, c='g')
axs[2].plot(tac, acor, lw=2, c='k')
axs[2].set_xlim(-5*tau, 5*tau)
axs[2].set_xlabel('$\Delta t$ (s)')
axs[2].set_ylabel('Correlation')
axs[2].set_title('Autocorrelation')
axs[2].legend(['True', 'Estimate'])

for ax in axs:
    set_font_size(ax, 16)
    
del t, xs, tau, ctr, ct, _t, dx, acor, tac, psd, frq, fig, axs
```
