# Power law emergence from random walk first passage times

Here we demonstrate that the first passage time for a random walk is distributed as a power-law. Specifically, we generate sequences of Poisson random variables where the sample at time t is used as the mean for the sample at time t+1. We plot a log-log distribution of how long it takes for the process to first reach zero and show that it approaches a straight line in the tail, indicating its power-law nature.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from disp import set_font_size


START = 4
N = 100000
MAX_ITER = 100000


ts = []
n_max = 0

for ctr in range(N):
    x = START
    for t_ctr in range(MAX_ITER):
        x = np.random.poisson(x)
        
        if x == 0:
            ts.append(t_ctr)
            break
            
    else:
        n_max += 1

# make hist
cts, bins = np.histogram(ts, bins=500)
bincs = .5 * (bins[:-1] + bins[1:])
        
fig, axs = plt.subplots(1, 2, figsize=(15, 7), tight_layout=True)

axs[0].plot(bincs, cts, lw=3)
axs[1].plot(np.log(bincs), np.log(cts), lw=3)

axs[0].set_xlabel('First zero time')
axs[0].set_ylabel('Occurrences')

axs[1].set_xlabel('Log[First zero time]')
axs[1].set_ylabel('Log[Occurrences]')

for ax in axs:
    ax.grid()
    set_font_size(ax, 16)
```
