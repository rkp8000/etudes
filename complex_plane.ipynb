```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np

from disp import set_font_size

r_ = np.linspace(-2, 2, 301)
j_ = np.linspace(-2j, 2j, 301)

r, j = np.meshgrid(r_, j_)
```

```python
z = 1/(2j - (r+j)**2) + 1/(-2j - (r+j)**2) + 1/(1 + (r + j)**2)

fig, ax = plt.subplots(1, 1, figsize=(10, 10), tight_layout=True)
ax.imshow(np.angle(z), vmin=-np.pi, vmax=np.pi, cmap='prism', extent=[r.min(), r.max(), r.min(), r.max()])
ax.set_xlabel('Real')
ax.set_ylabel('Imag')

set_font_size(ax, 16)
```

```python
z = 1/(-1-1j - (r + j))

fig, ax = plt.subplots(1, 1, figsize=(10, 10), tight_layout=True)
ax.imshow(np.angle(z), vmin=-np.pi, vmax=np.pi, cmap='prism', extent=[r.min(), r.max(), r.min(), r.max()])
ax.set_xlabel('Real')
ax.set_ylabel('Imag')

set_font_size(ax, 16)
```
